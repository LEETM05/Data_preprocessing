```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(bank_df['age'])
plt.xlabel('age')
plt.ylabel('freq')
plt.show()
```

```python
#연습문제 4
plt.hist(bank_df['balance'])
plt.xlabel('balance')
plt.ylabel('freq')
plt.show()

plt.hist(bank_df['day'])
plt.xlabel('day')
plt.ylabel('freq')
plt.show()

plt.hist(bank_df['duration'])
plt.xlabel('duration')
plt.ylabel('freq')
plt.show()

plt.hist(bank_df['campaign'])
plt.xlabel('campaign')
plt.ylabel('freq')
plt.show()
```

```python
plt.scatter(bank_df['age'], bank_df['balance'])
plt.xlabel('age')
plt.ylabel('balance')
plt.show()

print(bank_df[['age','balance']].corr())
```

```python
pd.plotting.scatter_matrix(bank_df[['age','balance','day','duration']])
plt.tight_layout()
plt.show()
```

```python
job_label = bank_df['job'].value_counts(ascending = False, normalize = True).index
job_vals = bank_df['job'].value_counts(ascending = False, normalize = True)
plt.pie(job_vals, labels = job_label)
plt.axis('equal')
plt.show()
```

```python
#6
print(bank_df['marital'].value_counts(ascending = False, normalize = True))

martial_label = bank_df['marital'].value_counts(ascending = False, normalize = True).index
martial_vals = bank_df['marital'].value_counts(ascending = False, normalize = True)

plt.pie(martial_vals, labels=martial_label)
plt.axis('equal')
plt.show()
```

```python
#6 
print(bank_df['education'].value_counts(ascending = False, normalize = True))

education_label = bank_df['education'].value_counts(ascending = False, normalize = True).index
education_vals = bank_df['education'].value_counts(ascending = False, normalize = True)

plt.pie(education_vals, labels=education_label)
plt.axis('equal')
plt.show()
```

```python
print(bank_df['y'].value_counts(ascending = False, normalize = True))

y_label = bank_df['y'].value_counts(ascending = False, normalize = True).index
y_vals = bank_df['y'].value_counts(ascending = False, normalize = True)

plt.pie(y_vals, labels=y_label)
plt.axis('equal')
plt.show()
```

```python
y_yes = bank_df[bank_df['y'] == 'yes']
y_no = bank_df[bank_df['y'] == 'no']
y_age = [y_yes['age'], y_no['age']]

plt.boxplot(y_age)
plt.xlabel('y')
plt.ylabel('age')
ax = plt.gca()
plt.setp(ax, xticklabels = ['yes', 'no'])
plt.show()
```
